```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy.stats as stats
```

```python
data=np.array([100,200,2000,300,20,1002,100,20000000000])
```

```python
data
```

```python
Z_score=np.abs(stats.zscore(data))
```

```python
threshold=3
```

```python
outliers=np.where(Z_score>threshold)[0]
```

```python
print(outliers)
```

```python
colors="red"
```

```python
plt.figure(figsize=(8,6))
plt.scatter(np.arange(len(data)),data, c=colors)
plt.xlabel("index")
plt.ylabel("price")
plt.title("finding outliers using z score method")
plt.show()
```
